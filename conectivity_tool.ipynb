```python
import urllib.request
import urllib.error

def check_website(url):
    if not url.startswith("http://") and not url.startswith("https://"):
        url = "https://" + url

    print(f"Checking status for {url}...")

    try:
        response = urllib.request.urlopen(url, timeout=5)
        if response.getcode() == 200:
            return "✅ The website is ONLINE and working perfectly!"
    except urllib.error.URLError as e:
        return f"❌ The website is DOWN or unreachable. Reason: {e.reason}"
    except Exception as e:
        return f"⚠ An unexpected error occurred: {e}"

if __name__ == "__main__":
    print("--- Website Connectivity Checker ---")
    user_url = input("Enter website URL (e.g., google.com): ").strip()
    status = check_website(user_url)
    print(status)
```
