# ***Fake News Generator***

```python
import random

subjects = [
    "Cat",
    "Robot",
    "Scientist",
    "Politician",
    "Alien",
    "Student",
    "Monkey"
]

actions = [
    "discovers",
    "creates",
    "destroys",
    "launches",
    "eats",
    "investigates",
    "reveals"
]

objects = [
    "a time machine",
    "a secret pizza recipe",
    "an invisible car",
    "a flying bicycle",
    "a chocolate planet",
    "a giant sandwich",
    "a new galaxy"
]

places = [
    "in New York",
    "on Mars",
    "inside a school",
    "under the ocean",
    "in a laboratory",
    "at the moon base"
]

print("=== Satirical News Generator ===\n")

headline = f"{random.choice(subjects)} {random.choice(actions)} {random.choice(objects)} {random.choice(places)}!"

print("Breaking News:")
print(headline)
```
