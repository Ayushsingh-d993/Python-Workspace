```python
# import time
import requests
from plyer import notification

# Config
COIN_ID = "bitcoin"  # Change to any crypto ID from CoinGecko
CURRENCY = "usd"
TARGET_PRICE = 60000  # Set your target alert price


def get_crypto_price():
    url = f"https://coingecko.com{COIN_ID}&vs_currencies={CURRENCY}"
    try:
        response = requests.get(url).json()
        return response[COIN_ID][CURRENCY]
    except Exception as e:
        print(f"Error fetching data: {e}")
        return None


def check_price():
    current_price = get_crypto_price()
    if current_price:
        print(f"Current {COIN_ID.capitalize()} price: ${current_price:,}")

        if current_price <= TARGET_PRICE:
            notification.notify(
                title=f"🚨 {COIN_ID.capitalize()} Price Alert!",
                message=f"Price dropped to ${current_price:,}! Target was ${TARGET_PRICE:,}.",
                timeout=10,
            )


if __name__ == "__main__":
    print("Crypto Tracker Started...")
    while True:
        check_price()
        time.sleep(300)  # Checks every 5 minutes
  # Checks every 5 minutes
```
